## K - Means

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
df = None


df = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/quantai/refs/heads/main/datasets/demographics.csv')
sns.scatterplot(x = df["bmi"], y = df["claim"], hue = KMeans(n_clusters=2).fit(df[["age","is_male","bmi","bloodpressure","diabetic","children","smoker","claim"]]).labels_).set(**{})
```

## TensorFlow

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd

train=pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/DSforBusiness/refs/heads/main/log_train.csv')

reglin = Sequential([Dense(1)])
reglin.compile(loss='mse')
reglin.fit(train[['time_page_seg',	'duration',	'sessions',	'pags_sess',	'conv']],train['value'])

test=pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/DSforBusiness/refs/heads/main/log_test.csv')
np.set_printoptions(precision=2, suppress=True)
reglin.predict(test[['time_page_seg',	'duration',	'sessions',	'pags_sess',	'conv']])
```
